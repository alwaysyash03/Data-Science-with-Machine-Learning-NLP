```python
### Read a whole file

with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
## Read a file line by line
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip())
```

```python
## Writing a file (overwritting)

with open('example.txt','w') as file:
    file.write('Hellow World\n')
    file.write('This is the new line.')
```

```python
## Write a file (without overwriting)
with open('example.txt', 'a') as file:
    file.write('\nAppend operation taking place.')
```

```python
## Writing multiple lines

line = ['\nfirst line', '\nsecond line', '\nthird line']

with open('example.txt', 'a') as file:
    file.writelines(line)
```

```python
## Binary Files
## Writing to a binary files

data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
    file.write(data)
```

```python
## Reading a binary file
with open('example.bin', 'rb') as file:
    content = file.read()
    print(content)
```

```python
## Read a content from source text file and write to a destination text file
with open('example.txt', 'r') as source_file:
    content = source_file.read()

with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)
```

```python
## Writing then Reading a file
with open('example.txt', 'w+') as file:
    file.write("Helloooo World\n")
    file.write("This is the new line.\n")

    ## Move the file cursor to the begining
    file.seek(0)
    
    ## Read the content of the file
    content = file.read()
    print(content)
```
